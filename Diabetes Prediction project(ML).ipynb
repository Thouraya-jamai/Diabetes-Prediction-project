# K-means


```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

k_values = [2, 3, 4]

for k in k_values:
    kmeans=KMeans(n_clusters=k , random_state=42)
    kmeans.fit(data_normalized)
   
    plt.figure(figsize=(6, 4))
    plt.scatter(data_normalized[:, 0],data_normalized[:, 1],c=kmeans.labels_,cmap='viridis',label="Classes assigned by K-means")
    plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:, 1], marker='x', c='red',label="cluster centers")
    plt.xlabel("Plasma glucose concentration")
    plt.ylabel("Blood pressure")
    plt.title(f'Clustering K-means with k={k}')
    plt.legend()
    plt.show()
    
```

# K-medoids

```python
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt 
k_values=[2,3,4]
for k in k_values:
    kmedoids=KMedoids(n_clusters=k,init='k-medoids++',random_state=42)
    kmedoids.fit(data_normalized)
    plt.scatter(data_normalized[:,0],data_normalized[:,1],c=kmedoids.labels_,cmap='viridis',label="Classes assigned by K-means")
    plt.scatter(data_normalized[kmedoids.medoid_indices_, 0], data_normalized[kmedoids.medoid_indices_, 1], marker='x', color='red', label='Medoïdes')
    plt.xlabel("Plasma glucose concentration")
    plt.ylabel("Blood pressure")
    plt.title(f'Clustering K-medoids with k={k}')
    plt.legend()
    plt.show()
```

# FCM

```python
import numpy as np
import skfuzzy as fuzz
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate test data
data, _ = make_blobs(n_samples=100, centers=3, cluster_std=1.0, random_state=42)

## Perform FCM
centers, u, _, _, _, _, _= fuzz.cluster.cmeans(data.T, c=3, m=2, error=0.005, maxiter=1000, init=None)
#display the Membership degree matrix:
print("Membership degree matrix:")
print(u)
```

```python
# Assign each data point to the cluster with the highest degree of membership
cluster_membership = np.argmax(u, axis=0)
# Show data points with colors corresponding to their cluster
for j in range(3):
    plt.scatter(data[cluster_membership == j, 0], data[cluster_membership == j, 1], label=f'Cluster {j+1}')
plt.scatter(centers[:, 0], centers[:, 1], marker='x', color='black', label='cluster centers')
plt.xlabel("Plasma glucose concentration")
plt.ylabel("Blood pressure")
plt.title('Result of categorization with FCM')
plt.legend()
plt.show()
```
